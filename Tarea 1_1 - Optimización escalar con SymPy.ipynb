# Optimización escalar con SymPy: criterio conocido

Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.

1. $f(x)=x+\frac{1}{x^2}$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = x + 1/x**2

df = sp.diff(f, x)
c = sp.solve(df, x)
print(c)
dff = sp.diff(df, x)
dff.subs(x, c[0])

fnum = sp.lambdify(x, f, 'numpy')
xvals = np.linspace(0.5,10, 100)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')
plt.plot((c[0]), fnum((c[0])), 'bd', label='Minimo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 1 y punto crítico')
plt.legend()
plt.grid(True)
plt.show()

print('El punto crítico es un minimo local')
```

2. $f(x)=1260+117x-9x^2$

```python
x = sp.symbols('x')
f = 1260 + 117*x - 9*x**2

df = sp.diff(f, x)
c = sp.solve(df, x)
print(c)
dff = sp.diff(df, x)
dff.subs(x, c[0])

fnum = sp.lambdify(x, f, 'numpy')
xvals = np.linspace(0, 10, 100)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')
plt.plot((c[0]), fnum((c[0])), 'bd', label='Maximo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 2 y punto crítico')
plt.legend()
plt.grid(True)
plt.show()

print('El punto crítico es un máximo local y global.')
```

3. $f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
x = sp.symbols('x', positive=True)
f = 5 + x + 4/x

df = sp.diff(f, x)
c = sp.solve(df, x)
print(c)
dff = sp.diff(df, x)
dff.subs(x, c[0])

fnum = sp.lambdify(x, f, 'numpy')
xvals = np.linspace(0.5, 10, 100)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')
plt.plot((c[0]), fnum((c[0])), 'bd', label='Mínimo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 3 y punto crítico')
plt.legend()
plt.grid(True)
plt.show()

print('El punto crítico es un mínimo local y global (Por ser positivo)')
```

4. $f(x)=\frac{15x}{2}\sqrt{\frac{x}{x-30}}$

```python
x = sp.symbols('x')
f = 15*x/2*sp.sqrt(x/(x-30))

df = sp.diff(f, x)
c = sp.solve(df, x)
print(c)
dff = sp.diff(df, x)
dff.subs(x, c[1])

fnum = sp.lambdify(x, f, modules=['numpy'])
xvals = np.linspace(38, 60, 100)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')
plt.plot(float(c[1]), fnum(float(c[1])), 'bd', label='Minimo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 4 y punto crítico')
plt.legend()
plt.grid(True)
plt.show()

print('El punto crítico es un minimo local y global (Por ser solo positivo)')
```

5. $f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
x = sp.symbols('x')
def f(x):
    return 6-2*x+sp.sqrt(18-2*x**2)
df = sp.diff(f(x), x)
c = sp.solve(df, x)
print(c)
dff = sp.diff(df, x)
dff.subs(x, c[0])

fnum = sp.lambdify(x, f(x), 'numpy')
xvals = np.linspace(-3, 3, 120)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')
plt.plot(float(c[0]), fnum(float(c[0])), 'bd', label='Minimo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 5 y punto crítico')
plt.legend()
plt.grid(True)
plt.show()
print('El punto crítico es un minimo local y global')
```

6. $f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
x = sp.symbols('x')
f = (1/4)*x**4 + (1/3)*x**3 - x**2

df = sp.diff(f, x)
c = sp.solve(df, x)
print(c)
dff = sp.diff(df, x)

fnum = sp.lambdify(x, f, 'numpy')
xvals = np.linspace(-3, 1.7, 200)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')

for cp in c:
    if float(cp) >= -3 and float(cp) <= 1.7:
        plt.plot(float(cp), fnum(float(cp)), 'rd', label=f'Crítico x={float(cp):.2f}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 6 y puntos críticos')
plt.legend()
plt.grid(True)
plt.show()

print('Clasificación de puntos criticos')
print('En x=-2 hay un minimo global, En x=0 hay un maximo local y en x=1 hay un minimo local')
```

7. $f(x)=100(1+\cos(x)\sin(x))$, para $x\in[0.\frac{\pi}{2}]$

```python
x = sp.symbols('x')
f = 100 * (1 + sp.cos(x) * sp.sin(x))

df = sp.diff(f, x)
c = sp.solve(df, x)
print("Puntos críticos:", c)
dff = sp.diff(df, x)

fnum = sp.lambdify(x, f, 'numpy')
xvals = np.linspace(0, np.pi/2, 200)
yvals = fnum(xvals)
plt.plot(xvals, yvals, label='f(x)')
plt.plot(float(c[1]), fnum(float(c[1])), 'bd', label='Máximo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función 7 y puntos críticos')
plt.legend()
plt.grid(True)
plt.show()

print('El punto crítico es un máximo local y global (Por el intervalo dado)')
```
